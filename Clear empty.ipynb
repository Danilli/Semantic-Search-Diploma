```python
import json
import os
```

```python
#if there no text into json file we should to delete it
j = 0
for i in range(1, 1805):
    try:
        file_path = f"{i}.json"
        with open(file_path, "r") as f:
            data = json.load(f)
            if data["text"] == "":
                j+=1
                print(f"{i} file was deleted")
                f.close()
                os.remove(file_path)
    except OSError as e:
        print("Oops...", e)
print(f"\ndeleted {j} files")
```
